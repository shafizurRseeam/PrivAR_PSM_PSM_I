```python


def generate_plm_samples(epsilon: float, num_samples: int) -> np.ndarray:
    """Vector of PLM radii."""
    return np.array([sample_laplace_radius(epsilon) for _ in range(num_samples)], dtype=float)

def generate_psm_samples(epsilon: float, num_samples: int, delta: float = 1.0) -> np.ndarray:
    """Vector of PSM radii."""
    return np.array([sample_psm_radius(epsilon, delta=delta) for _ in range(num_samples)], dtype=float)    
```

```python

if __name__ == "__main__":
    epsilon = 0.1
    delta = 1.0
    num_samples = 100_0000

    rs = generate_psm_samples(epsilon, num_samples, delta=delta)  # name changed
    rl = generate_plm_samples(epsilon, num_samples)

    print("PLM radius: mean = {:.3f}, max = {:.3f}, sum = {:.3f}".format(rl.mean(), rl.max(), rl.sum()))
    print("PSM radius: mean = {:.3f}, max = {:.3f}, sum = {:.3f}".format(rs.mean(), rs.max(), rs.sum()))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity

#  Font size settings
label_fontsize = 28
tick_fontsize = 21
legend_fontsize = 28
title_fontsize = 22

# Your data
rs_array = np.array(rs).reshape(-1, 1)
rl_array = np.array(rl).reshape(-1, 1)

# # Fit KDE models
# kde_rs = KernelDensity(kernel='gaussian', bandwidth=0.1).fit(rs_array)
# kde_rl = KernelDensity(kernel='gaussian', bandwidth=0.6).fit(rl_array)

kde_rs = KernelDensity(kernel='gaussian', bandwidth=0.1).fit(rs_array)
kde_rl = KernelDensity(kernel='gaussian', bandwidth=0.1).fit(rl_array)

# Define a common x-axis range
x_min = min(rs_array.min(), rl_array.min())
x_max = max(rs_array.max(), rl_array.max())

# Create full x grid
x_vals_full = np.linspace(x_min, x_max, 1000).reshape(-1, 1)

# Evaluate densities
pdf_rs_full = np.exp(kde_rs.score_samples(x_vals_full))
pdf_rl_full = np.exp(kde_rl.score_samples(x_vals_full))

# Filter to x <= 60
x_cutoff = 60
mask = x_vals_full.flatten() <= x_cutoff
x_vals = x_vals_full.flatten()[mask]
pdf_rs = pdf_rs_full[mask]
pdf_rl = pdf_rl_full[mask]
#plt.figure(figsize=(10, 8))  # Increased height for clarity

# Plot both densities with fill
plt.fill_between(x_vals, pdf_rs, color='royalblue', alpha=0.9, label='PSM')
plt.fill_between(x_vals, pdf_rl, color='salmon', alpha=0.9, label='PLM')

plt.xlabel('Distance (meters)', fontsize=label_fontsize)
plt.ylabel('Density', fontsize=label_fontsize)
plt.tick_params(axis='both', which='major', labelsize=tick_fontsize)
plt.legend(fontsize=legend_fontsize)
plt.tight_layout()

# Save as both SVG and PDF
output_base = r'C:\Users\ss6365\Desktop\PrivAR_PSM_PSM_I\figures\psm_plm_pdf'
# #plt.savefig(output_base + '.svg', format='svg', transparent=True)
plt.savefig(output_base + '.pdf', format='pdf', transparent=True)

plt.show()
```

```python
import time
import numpy as np

epsilon = 0.1
iters = 200_000  # enough to stabilize

# warmup
_ = sample_laplace_radius(epsilon)
_ = sample_psm_radius(epsilon, delta=1.0)

t0 = time.perf_counter()
for _ in range(iters):
    _ = sample_laplace_radius(epsilon)
t1 = time.perf_counter()

t2 = time.perf_counter()
for _ in range(iters):
    _ = sample_psm_radius(epsilon, delta=1.0)
t3 = time.perf_counter()

print("PLM per-sample:", (t1 - t0)/iters)
print("PSM per-sample:", (t3 - t2)/iters)
```
